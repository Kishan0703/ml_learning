using xgbregerssor

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.datasets import fetch_california_housing
# from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
data=fetch_california_housing()
```

```python
df= pd.DataFrame(data.data, columns=data.feature_names)
```

```python
df['price']=data.target
```

```python
df
```

```python
df.describe()
```

```python
df.var()
```

splitting data

```python
X=df.drop('price',axis=1)
Y=df['price']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2, random_state=42)
```

training model

```python
# model=LinearRegression()
```

```python
model=XGBRegressor()
```

```python
model.fit(X_train,Y_train)
```

```python
# training data
X_prediction=model.predict(X_train)
```

```python
X_prediction
```

```python
r2val1=metrics.r2_score(X_prediction,Y_train)
ma_error1=metrics.mean_absolute_error(X_prediction,Y_train)
```

```python
print(r2val1)
print(ma_error1)
```

```python
# testing data
X_test_predict=model.predict(X_test)
```

```python
r2val2=metrics.r2_score(X_test_predict,Y_test)
ma_error2=metrics.mean_absolute_error(X_test_predict,Y_test)
```

```python
print(r2val2)
print(ma_error2)
```
