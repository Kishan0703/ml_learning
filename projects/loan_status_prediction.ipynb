```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
data= pd.read_csv('/Users/kishan/code/genai/ml/ml_learning/datasets/loan_data.csv')
```

```python
data.dropna(inplace=True)
```

```python
data.head()
```

```python
data.replace({'Loan_Status':{"N":0, "Y":1}}, inplace=True)
```

```python
data.head()
```

```python
data.Dependents.value_counts()
```

```python
data.replace(to_replace='3+',value=4,inplace=True)
```

```python
data.Dependents.value_counts()
```

```python
data.replace({'Gender':{'Male':1,'Female':0},'Married':{'Yes':1,'No':0},'Education':{'Graduate':1,'Not Graduate':0},'Self_Employed':{'Yes':1,'No':0},'Property_Area':{'Rural':0,'Urban':1,'Semiurban':0.5}},inplace=True)
```

```python
data.head()
```

```python
data.groupby('Loan_Status').size()
```

```python
X=data.drop(columns=['Loan_ID','Loan_Status'],axis=1)
Y=data.Loan_Status
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2,stratify=Y)
```

```python
model=svm.SVC(kernel='linear')
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_predict=model.predict(X_train)
X_train_predict_accuracy=accuracy_score(X_train_predict,Y_train)
print(X_train_predict_accuracy)
```

```python
X_test_predict=model.predict(X_test)
X_test_predict_accuracy=accuracy_score(X_test_predict,Y_test)
print(X_test_predict_accuracy)
```
