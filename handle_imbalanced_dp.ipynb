```python
from sklearn.datasets import fetch_openml

credit = fetch_openml("credit-g", version=1, as_frame=True)
df = credit.frame

print(df['class'].value_counts())
```

```python
df
```

```python
good=df[df['class']=='good']
bad=df[df['class']=='bad']
```

```python
print(good.shape, bad.shape)
```

```python
reduced_good=good.sample(300)
reduced_good.shape
```

```python
new_df=pd.concat([reduced_good,bad],axis=0)
```
